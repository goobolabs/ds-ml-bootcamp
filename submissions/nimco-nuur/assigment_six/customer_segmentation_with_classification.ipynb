```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)
import pandas as pd
```

```python
# --------------------------------
# Load clustered dataset
# --------------------------------
INPUT_PATH = "dataset/segmented_wholesale_customers.csv"
clustered_df = pd.read_csv(INPUT_PATH)


print("\n=== LOADED CLUSTERED DATASET ===")
print(clustered_df.head())

print("\n=== DATASET INFO ===")
clustered_df.info()
```

```python

# Features
FEATURES = [
    "Fresh",
    "Milk",
    "Grocery",
    "Frozen",
    "Detergents_Paper",
    "Delicassen"
]

# X = Features
X = clustered_df[FEATURES]

# y = Cluster labels
y = clustered_df["Cluster"]

```

```python
# Cluster Classification
# --------------------------------


# Features
X_cls = clustered_df[FEATURES]

# Target = K-Means cluster labels
y_cls = clustered_df["Cluster"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_cls,
    y_cls,
    test_size=0.2,
    random_state=42,
    stratify=y_cls
)
```

```python
# Train classifier
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

```

```python
rf.fit(X_train, y_train)

# Predict
y_pred = rf.predict(X_test)

print("\n=== RANDOM FOREST EVALUATION ===")

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average="weighted")
recall = recall_score(y_test, y_pred, average="weighted")
f1 = f1_score(y_test, y_pred, average="weighted")

print(f"Accuracy : {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall   : {recall:.3f}")
print(f"F1-Score : {f1:.3f}")

cm = confusion_matrix(y_test, y_pred)

cm_df = pd.DataFrame(
    cm,
    index=[f"Actual {i}" for i in range(5)],
    columns=[f"Predicted {i}" for i in range(5)]
)

print("\n=== CONFUSION MATRIX ===")
print(cm_df)
```

```python
# --------------------------------
# 11. Predict a New Customer
# --------------------------------
new_customer = pd.DataFrame([{
    "Fresh": 15000,
    "Milk": 5000,
    "Grocery": 7000,
    "Frozen": 3000,
    "Detergents_Paper": 1200,
    "Delicassen": 900
}])

predicted_cluster = rf.predict(new_customer)

print("\n=== NEW CUSTOMER PREDICTION ===")
print(f"Predicted Cluster: {predicted_cluster[0]}")
```
