```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

Part B — Practical: Car Price Prediction

1.Load Dataset

```python
CSV_PATH="clean_car_dataset.csv"
df=pd.read_csv(CSV_PATH)
print(df.head())
```

2-Prepare Features & Target

```python
x=df.drop(columns=["Price", "LogPrice"])
y=df["Price"]

```

3-Split Data

```python
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

4-Train Models

Train Linear Regression Predivtions 

```python
lr=LinearRegression()
lr.fit(x_train, y_train)
lr_pred=lr.predict(x_test)
```

```python
print("Linear Regression Predictions: ")
print(lr_pred)
```

Train RandomForestRegressor

```python
rf=RandomForestRegressor(n_estimators=100, random_state=42 )
```

```python
rf.fit(x_train, y_train)
rf_pred=rf.predict(x_test)
```

```python
print("Random Forest Predictions: ")
print(rf_pred)
```

5-Evaluate Performance

```python
def print_metrics(name,y_true, y_pred):
    r2=r2_score(y_true, y_pred)
    mae= mean_absolute_error(y_true, y_pred)
    mse=mean_squared_error(y_true, y_pred)
    rmse=np.sqrt(mse)
    print(f"{name} Performance:")
    print(f"  R²   :{r2:.3f} ")
    print(f" MAE   : {mae:,.0f}")
    print(f" MSE   : {mse:,.0f}")
    print(f" RMSE  : {rmse:,.0f}")
        
    
```

```python
print_metrics("Linear Regression" , y_test,lr_pred)
print_metrics("Random Forest", y_test, rf_pred)
```

6-Sanity Check

```python
i=3
x_one_df= x_test.iloc[[i]]
y_true=y_test.iloc[i]
```

```python
p_lr_one=float(lr.predict(x_one_df)[0])
p_rf_one=float(rf.predict(x_one_df)[0])

```

```python
print("Single-row sanity check: ")
print(f"Actual Price : ${y_true: ,.0f}")
print(f"Linear Regression Predictions : ${p_lr_one: ,.0f}")
print(f"Random Forest Prediction : ${p_rf_one: ,.0f}")

```
